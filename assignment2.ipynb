```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
df_sal = pd.read_csv('C:/Users/harsh/Downloads/Salary_Data.csv')
df_sal.head()
```

```python
df_sal.describe()
```

```python
plt.scatter(df_sal['YearsExperience'],df_sal['Salary'],color='purple')
plt.title('salary vs experience')
plt.xlabel('year of exp')
plt.ylabel('salary')
plt.show()
```

```python
x=df_sal.iloc[:,:1] #experience
y=df_sal.iloc[:,1:] #salary
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred_train=model.predict(x_train)
y_pred_test=model.predict(x_test)
```

```python
plt.scatter(x_train, y_train, color = 'purple')
plt.plot(x_train, y_pred_train, color = 'black')
plt.title('Salary vs Experience for Training')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x_test, y_test, color = 'purple')
plt.plot(x_train, y_pred_train, color = 'black')
plt.title('Salary vs Experience for Testing')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
print(f'Coefficient: {model.coef_}')
print(f'Intercept: {model.intercept_}')
```

```python
from sklearn.metrics import mean_squared_error
print('Mean square error(test):', mean_squared_error(y_test,y_pred_test))
```

```python
from sklearn.metrics import mean_absolute_error
print('Mean absolute error(test):',mean_absolute_error(y_test,y_pred_test))
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred_test)
print('r2 score:', r2)
```
